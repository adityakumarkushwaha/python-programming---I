( i ) . Print even numbers between 1 and 50 .

( ii ) . skip numbers that are mutiplies of 10 using continue statement .

( iii ) . stop the loop  if  a number greater than 40 is found using break .


```python
# program to print even numbers between 1 and 50

# skip multiples of 10

# stop if number > 40

for x in range(1,51):
    if x % 2 != 0:
        continue
    if x % 10 == 0:
        continue
    if x > 40:
        break
    print (x)

```

( i ) . palindrome check

```python
x = int(input("Enter a number:"))
if str (x) == str (x)[::-1]:
    print ("The number is a palindrome")
else:
    print ("The number is not a palindrome")
```

( ii ) . Armstrong number check

```python
x = int(input("Enter a number:"))
sum = 0
temp = x
while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10
if x == sum:
    print (x, "is an Armstrong number")
else:
    print (x, "is not an Armstrong number")
```
