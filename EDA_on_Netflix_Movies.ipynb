## Distribution of Movies vs TV Shows

```python
sns.countplot(x='type', data=netflix, palette='coolwarm')
plt.title("Movies vs TV Shows")
plt.show()
```

 ## Content Added by Year

```python
plt.figure(figsize=(22,13))
sns.countplot(data=netflix,x='release_year')
plt.title("Titles per Year")
plt.xticks(rotation= 90)
plt.show()
```

## Top Content Producing Countries

```python
country_data = netflix['country'].dropna().str.split(',').explode().value_counts().head(10)


plt.figure(figsize=(10,4))
sns.barplot(x=country_data.index, y=country_data.values)
plt.title("Top 10 Content Producing Countries")
plt.xticks(rotation=65)
plt.show()
```

## Release Year Trend

```python
plt.figure(figsize=(12,4))
sns.histplot(netflix['release_year'], bins=30, kde=True)
plt.title("Release Year Distribution")
plt.show()
```

## Genre in Analysis

```python
genres = netflix['listed_in'].str.split(',').explode().value_counts().head(15)

plt.figure(figsize=(10,4))
sns.barplot(x=genres.values, y=genres.index)
plt.title("Top Genres on Netflix")
plt.show()
```

## Duration Analysis

```python
movies = netflix[netflix['type']=='Movie']
movies['duration_mins'] = movies['duration'].str.replace(" min","").astype(float)

plt.figure(figsize=(10,4))
sns.histplot(movies['duration_mins'], bins=30, kde=True)
plt.title("Movie Duration Distribution")
plt.show()
```

## TV Show Seasons

```python
tv = netflix[netflix['type']=='TV Show']
tv['seasons'] = tv['duration'].str.replace(" Season","").str.replace("s","").astype(int)

sns.countplot(x='seasons', data=tv)
plt.title("TV Show Seasons Count")
plt.show()
```

## Rating Analysis

```python
plt.figure(figsize=(10,4))
sns.countplot(y='rating', data=netflix, order=netflix['rating'].value_counts().index)
plt.title("Content Rating Distribution")
plt.show()
```
