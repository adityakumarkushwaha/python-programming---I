```python
plt.figure(figsize=(6,4))
plt.hist(df['Profit'],bins =40)
plt.title("Super_Market Profit ")
plt.xlabel(" Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['Profit'])
plt.title('Sales boxplot')
plt.show()
```

```python
plt.scatter(df['Sales'],df['Discount'])
plt.title('Sales Vs Discount ')
plt.xlabel("Sales")
plt.ylabel("Discount")
plt.show()
```

```python
sns.barplot(data=df,y='Sales',x='Ship Mode')
plt.title("SALES Vs SHIP MODE")
plt.xlabel("SALES")
plt.ylabel("SHIP MODE")
plt.show()
```

```python
sns.countplot(data=df,x='Ship Mode')
plt.title("SHIPING MODE")
plt.show()
```

```python
sns.histplot(df['Ship Mode'],kde=True)
plt.title("SHIPING MODE ")
plt.show()
```
