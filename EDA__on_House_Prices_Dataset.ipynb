## Histogram

```python
plt.figure(figsize=(10,5))
plt.hist(df['price'], bins=30)
plt.title("House Price Distribution")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['price'],kde=True)
plt.show()
```

## Boxplot 

```python
plt.boxplot(df['price'])
plt.title("SalePrice Boxplot")
plt.show()
```

```python
sns.boxplot(data=df,x='bedrooms',y='floors')
plt.show()
```

## Scatter plot

```python
plt.scatter(df['floors'], df['price'])
plt.xlabel("Floor Area")
plt.ylabel("Sale Price")
plt.title("Floor Area vs Price")
plt.show()
```

## Bar plot

```python
plt.figure(figsize=(10,4))
sns.barplot(data = df, x='bedrooms', y='price')
plt.title("Bedroom vs Price")
plt.show()
```

## Count Plot

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='floors' )
plt.title('Counting Floors')
plt.show()
```

## Violin Plot

```python
sns.violinplot(data=df,x='floors',y='bathrooms')
plt.show()
```

## Pair Plot

```python
sns.pairplot(df[['view','bedrooms','bathrooms','grade']])
plt.show()
```
