```python
plt.figure(figsize=(8,4))
sns.countplot( data=Wine,x='quality')
plt.title("Wine Quality Count")
plt.show()
```

```python
Wine.hist(figsize=(14,10), bins=20)
plt.tight_layout()
plt.show()
```

```python
for col in Wine.columns[:-1]:
    plt.figure(figsize=(6,4))
    sns.histplot(Wine[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
corr = Wine.corr(numeric_only=True)

plt.figure(figsize=(12,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='quality', y='alcohol', data=Wine)
plt.title("Alcohol vs Quality")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='quality', y='volatile acidity', data=Wine)
plt.title("Volatile Acidity vs Quality")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='quality', y='sulphates', data=Wine)
plt.title("Sulphates vs Quality")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=Wine)
plt.xticks(rotation=90)
plt.title("Outlier Detection")
plt.show()
```
