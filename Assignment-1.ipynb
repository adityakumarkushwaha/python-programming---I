# Interchange list

```python
# Python program to interchange list
my_list = [10, 20, 30, 40, 50]
print("Original list:", my_list)

# --- Method 1: Using a Temporary Variable ---
temporary = my_list[0]
my_list[0] = my_list[-1]
my_list[-1] = temporary

print("After swapping (using temporary variable):", my_list)

# --- Method 2: Without Using Temporary Variable ---
# Swapping again to original using tuple unpacking
my_list[0], my_list[-1] = my_list[-1], my_list[0]

print("After swapping (without using temporary variable):", my_list)
```
