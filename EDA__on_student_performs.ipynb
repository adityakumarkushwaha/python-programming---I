```python
cols = ["Nota_1Semestre","Nota_2Semestre","Trabalho_Pai","idade","Sexo"]
sns.pairplot(df[cols],hue="Sexo")
plt.show()
```

```python
cols = ["Nota_1Semestre","Nota_2Semestre","Trabalho_Pai","idade","Sexo"]

corr = df[cols].corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Selected Feature Correlation Heatmap")
plt.show()
```
