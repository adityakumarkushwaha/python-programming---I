```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(df['sepal_width'],kde = 1)
plt.title("Figure_width")
plt.show()
```

```python
sns.countplot(data=df,x='species')
plt.title("Species Count")
plt.show()
```

```python
sns.boxplot(data=df, x='species')
plt.title("Species Boxplot")
plt.show()
```

```python
sns.scatterplot(data=df,x='sepal_length',y='sepal_width')
plt.title("Sepal_length vs Sepal_width")
plt.show()
```

```python
sns.barplot(data = df, y='petal_length' , x='species')
plt.title("Petal_length vs Species")
plt.show()
```

```python
sns.boxplot( data=df,x='species',y='sepal_width')
plt.title("Sepal_width by Species ")
plt.show()
```

```python
sns.pairplot(df[['sepal_length','sepal_width','petal_length','petal_width']])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.violinplot( data=df,x='species',y='sepal_width')
plt.title("Sepal_width by Species ")
plt.show()
```

```python
sns.swarmplot(data=df,x='sepal_length',y='species')
plt.title("Fare Distribution by Class")
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', hue='species', fill=True)
plt.title("sepal_lenght by Species")
plt.show()
```

```python
df['sepal_width'].plot(kind='hist', bins=30, title='Sepal_width Histogram')
plt.show()
```

```python

df.plot(kind='box', subplots=True, layout=(3,4), figsize=(12,8))
plt.show()
```
